```python
import matplotlib.pyplot as plt
import numpy as np

# plot configuration
# ==================
plt.figure(figsize=(10,10))
# plt.grid()

# range
plt.ylim(ymin=-20) 
plt.ylim(ymax=20) 

# domain
# ======
x = np.linspace(-10,10,1000)

# collection of plots
plots = []

# create tangent line
# ===================
def D(f,x,dx=1E-4): # derivative
    return ( f(x + dx) - f(x) )/dx

def tangent(f,c): # line
    return D(f,c)*(x-c)+f(c)

# functions
# =========
def f(x):
    return x**2


# tangent line plotter
# ====================
def plotTan(f,a,b):
    for c in np.arange(a,b+1):    
        plots.append( plt.plot(x,tangent(f,c)) )
        
# plot function
plots.append( plt.plot(x,f(x)) )
# plot function tangent lines
plotTan(f,-100,100)

# display plot
plt.show()
```
