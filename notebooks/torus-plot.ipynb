```python
def g(x):
    return x**3

def norm(vector,t):
    sum = 0
    
    for i in vector:
        sum += i**2
        
    return sqrt(sum)

# unit tangent
def T(vector,t):
    tangent = []
    
    diffVector = []
    for f in vector:
        diffVector.append( diff(f,t) )
    
    n = norm(diffVector,t)
    
    for f in vector:
#         tangent.append( lambdify(t, diff(f,t)/n) )
        tangent.append( diff(f,t)/n )
        
    return tangent

func = [cos(t),sin(t)]

lambdify(t, T(func,t))(2)

# unit normal
lambdify(t, T(T(func,t),t))(2)
```
