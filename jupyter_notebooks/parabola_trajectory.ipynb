```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,10))
plt.grid()
MAX = 15 # max range in both directions
x = np.linspace(-MAX,MAX,1000) # x range

def f(x):
    return x**2+b*x # quadratic

plots = []
for b in np.linspace(-MAX,MAX,MAX):
    h = -b/2 # vertex x
    k = f(h) # vertex y
    
    plots.append( plt.plot(x,f(x),color='blue') ) # plot parabola
    plots.append( plt.plot(h,k,'ro') ) # plot vertex

# plot the additive inverse parabola
plots.append(plt.plot(x,-x**2,color='red'))

plt.show()
```
