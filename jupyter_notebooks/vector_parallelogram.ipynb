```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

MAX = 10
plt.figure(figsize=(10,10))
plt.xlim(0,MAX)
plt.ylim(0,MAX)
plt.grid()

ax = plt.axes()

v = np.array([[2,3],[5,2]]) # vectors
colors = ['b','r']

i = 0
for vector in v:    
    ax.arrow(0, 0, vector[0], vector[1], ec=colors[i], fc=colors[i])
    i += 1
    
ax.arrow(0,0, v[0,0]+v[1,0],v[0,1]+v[1,1],color='purple') # vector sum

ax.arrow(v[1,0],v[1,1], v[0,0],v[0,1],color='b') # vector translation 1
ax.arrow(v[0,0],v[0,1], v[1,0],v[1,1],color='r') # vector translation 2

plt.show()
```
