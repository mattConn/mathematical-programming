# The Parabolic Trajectory of The Parabola's Vertex
$ax^2+bx+c$  

As the $b$ term changes in the quadratic equation while the $a$ and $c$ term remain constant, a parabolic trajectory can be seen, which would trace a plot of the additive inverse of the given quadratic equation.  

The above is demonstrated below via 20 (an arbitrary even value) plotted parabolas with whose $b$ terms range from -20 to 20 each.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

x = np.linspace(-20,20)

for b in range(-20,21):
    # vertex x and y
    h = -b/2
    k = h**2+b*h
    
    # parabolas (no y intercept)
    plt.plot(x, x**2+b*x, color='blue', linewidth=1)
    #vertices
    plt.plot([h], [k], marker='o', markersize=3, color="red")

# plot the additive inverse
plt.plot(x, -x**2, color='red', linewidth=0.5)

# set range of plot to be displayed
plt.ylim(-120,100)
plt.xlim(-15,15)
plt.grid()
```
