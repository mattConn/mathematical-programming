```python
import matplotlib.pyplot as plt
from numpy import *

x1 = np.linspace(0,2*pi,1000)
x2 = np.linspace(0,1,1000) # restricted domain for inverse

functions = [
    [
    sin(x1),
    cos(x1),
    tan(x1)
    ],
    
    # recprocals
    [
    1/sin(x1),
    1/cos(x1),
    1/tan(x1)
    ],
    
    # inverse
    [
    arcsin(x2),
    arccos(x2),
    arctan(x2)
    ]
    
]

plt.figure(figsize=(10,5))

f1 = plt.gca()
f1.set_ylim(-1,1)
f1.grid()


for i in functions[0]:
    f1.plot(x1,i)
    
f2 = plt.gca()
f2.set_ylim(-1,1)
f2.grid()

    
for i in functions[1]:
    f2.plot(x1,i)
```
