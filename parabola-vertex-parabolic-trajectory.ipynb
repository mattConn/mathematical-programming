```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

x = np.linspace(-20,20)

for i in range(-20,21):
    # vertex x and y
    h = -i/2
    k = h**2+i*h
    
    # parabolas (no y intercept)
    plt.plot(x,x**2+i*x,color='blue',linewidth=1)
    #vertices
    plt.plot([h], [k], marker='o', markersize=3, color="red")

plt.ylim(-120,100)
plt.xlim(-15,15)
plt.grid(True)
```
