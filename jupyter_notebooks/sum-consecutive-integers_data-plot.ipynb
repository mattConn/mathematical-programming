```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

file = open("../sum_consecutive_integers/sum-consecutive-integers_10^6.txt","r")

MAX = 10**6

plt.ylim(1,90)
plt.xlim(1,10)
x = np.arange(1,MAX+1)
l = [] #list of sums

for line in iter(file):
    l.append(line)

plt.grid()
plt.plot(x,l,'bo-',label="1+2+3+...")
plt.plot(x,x**2,'ro-',label="x^2")
plt.legend()
```
